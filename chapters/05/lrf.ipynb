# Low Rank Factorization

```python
import sys
sys.path.insert(0, '../../../molass-library')
from molass_data import TUTORIAL_DATA
print(TUTORIAL_DATA)
from molass.DataObjects import SecSaxsData as SSD
ssd = SSD(TUTORIAL_DATA)
ssd.plot_3d();
```

```python
trimmed_ssd = ssd.trimmed_copy()
corrected_ssd = trimmed_ssd.corrected_copy()
decomposition = corrected_ssd.quick_decomposition()
plot1 = decomposition.plot_components() 
```

```python
decomposition3 = corrected_ssd.quick_decomposition(num_components=3)
plot2 = decomposition3.plot_components() 
```

```python
proportions = decomposition3.get_proportions()
proportions
```

```python
decomposition3 = corrected_ssd.quick_decomposition(num_components=3, proportions=[0.57, 0.10, 0.33])
plot2 = decomposition3.plot_components() 
```
