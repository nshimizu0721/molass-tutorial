```python
import sys
sys.path.insert(0, r"D:\Github\molass-legacy") 
sys.path.insert(0, r"D:\Github\molass-library")
import numpy as np
import matplotlib.pyplot as plt
from molass.Local import get_local_settings
from molass.DataObjects import SecSaxsData as SSD
local_settings = get_local_settings()
DATA_ROOT_FOLDER  = local_settings['DATA_ROOT_FOLDER']
path = DATA_ROOT_FOLDER + r'\20160227\backsub'
ssd = SSD(path)
ssd.plot_3d();
```

```python
ssd.plot_compact();
```

```python
from molass.Stats.EghMoment import EghMoment
uv_icurve = ssd.uv.get_icurve()
uv_moment = EghMoment(uv_icurve)
xr_icurve = ssd.xr.get_icurve()
xr_moment = EghMoment(xr_icurve)
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
uv_moment.debug_plot(ax1)
xr_moment.debug_plot(ax2)
fig.tight_layout()
```
